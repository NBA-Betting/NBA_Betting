```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
username = 'postgres'
password = ''
endpoint = ''
database = 'nba_betting'
port = '5432'
```

```python
connection= create_engine(f'postgresql+psycopg2://{username}:{password}@{endpoint}/{database}').connect()
```

```python
df = pd.read_csv("2022-04-09T21-48-00.csv")
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.head()
```

```python
mask = (df['date'] >= "2022-02-28") & (df['date'] <= "2022-04-08")
```

```python
df.loc[mask]
```

```python
needed_data = df.loc[mask]
```

```python
needed_data.info()
```

```python
needed_data["id_num"] = needed_data["game_id"]
needed_data["link"] = needed_data["game_url"]
needed_data["home_team_short_name"] = needed_data["team"]
needed_data["away_team_short_name"] = needed_data["opponent"]
needed_data["open_line_home"] = needed_data["spread"]
needed_data["open_line_away"] = - needed_data["spread"]
```

```python
needed_data = needed_data.drop(columns=['home', 'game_id', 'game_url', 'team', 'opponent', 'score', 'opponent_score', 'result', 'spread_result', 'spread'])
```

```python
needed_data["date"] = needed_data['date'].apply()
```

```python
needed_data.info()
```

```python
needed_data
```

```python
needed_data.to_sql("covers_odds", connection, if_exists="append", index=False)
```
